```python
import pandas as pd 
import numpy as np
from datetime import datetime

df= pd.read_csv('company.csv') 
df.dropna(inplace=True ) # Poum retire ranje ki vid yo
df.dropna(axis=1, inplace=True) # Poum retirer kolon ki manke vale 

#Poum chanje 'K' ki nan kolon salaire lan
def konveti_K(salary):
    if isinstance(salary, str) and salary.endswith('K'):
        return (salary[:-1]) + '000'
    return (salary)
df['salary'] = df['salary'].apply(konveti_K) 

#Poum ka retire vigil yo, et apres poum metel en tip float 
df['salary'] = df['salary'].str.replace(',','').astype(float)

#Poum retire tout years ki nan years_at_company an 
df['years_at_company'] = df['years_at_company'].str.replace('years','').astype(int) 

#Poum chanje tout Male en M, et tout female en F 
df['gender'] = df['gender'].replace({'Male':'M', 'Female':'F'})

#Poum change format date yo 
fomatdat = ['%d-%m-%Y', '%d-%b-%Y', '%Y-%m-%d', '%d-%m-%y', '%m/%d/%Y', '%d-%b-%y']
def konveti_dat(dat):
    for elem in fomatdat:
        try:
            return pd.to_datetime(dat, format=elem, dayfirst=True)
        except ValueError:
            continue
    return pd.NaT 
df['hired_date'] = df['hired_date'].apply(konveti_dat)

#Kesyon sou ANALIZ DEMOGRAFIK
#Repatisyon chak anplwaye nan chak depatman.
print(      '***ANALIZ DEMOGRAFIK***        ')
repdepatman = df['department'].value_counts()
print('Repatisyon pa Depatman:')
print(repdepatman)
print(     )

#Repatisyon pa sèks nan chak depatman.
repsex = df.groupby(['department', 'gender']).size().unstack(fill_value=0)
print('Repatisyon pa Seks:')
print(repsex)
print(     )

#Poum ka we kantite employe ki menm laj par departement. 
lajdep = df.groupby('department')['age'].value_counts()
print(lajdep)
print(     )

#Mwayèn laj anplwaye yo pou chak depatman.
moylaj = df.groupby('department')['age'].mean()
print('Mwayen Laj Pa Depatman:')
print(moylaj)
print(   )

print(      '***SALE***        ')
#Poum ka we kantite employe ki menm salaire par departement. 
totalsale = df.groupby('department')['salary'].value_counts()
print(totalsale)
print(     )

#Mwayèn salè nan chak depatman.
moysale = df.groupby('department')['salary'].mean()
print(moysale)
print(   )

# Satisfaksyon travay sou chak depatman.
totalsatisfaksyon= df.groupby('department')['job_satisfaction'].value_counts()
print(totalsatisfaksyon)
print(   )

print(      '*** PWOMOSYON AK EDIKASYON***        ')
#Mwen use fonction sa poum ka cherche intervall de temps ki genyen depi denye promotion an te fet lan.
def intevaltan(df):
    current_year = datetime.now().year
    df['years_since_last_promotion'] = current_year - df['last_promotion_year']
    return df 

#Mwayèn tan ki genyen depi dènye fwa konpayi a te bay yon pwomosyon nan chak depatman.
def moy_denye_promosyon(df):
    df = intevaltan(df)
    return df.groupby('department')['years_since_last_promotion'].mean()

#Poum rele fonksyon an
moyenne_tan = moy_denye_promosyon(df)
print(moyenne_tan)
print(     )

#Mwayèn salè ki ekziste an fonksyon de nivo edikasyon anplwaye yo (Bachelor, Master.
moySale = df.groupby('education_level')['salary'].mean()
print(moySale)
print(     )

df
```
